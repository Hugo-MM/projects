### This notebook was developped in order to process the synopsis of all movies into a csv file that could  be fed into the app database thanks to a django command.

```python
import numpy as np
import pandas as pd
import csv
```

```python
all_movies = pd.read_csv('archive/IMDB movies.csv',
                        usecols= [
                            'imdb_title_id',
                            'description'
                        ])
```

```python
all_movies.info()
```

```python
all_movies = all_movies.fillna('')
```

```python
movie_data = pd.read_json('data/data_for_app/movie_data.json')
```

```python
movie_data['imdb_title_id'] = movie_data['fields'].apply(lambda x : x['imdbid'])
```

```python
movie_data = movie_data.drop(['fields', 'model'], axis=1)
```

```python
movie_data.head()
```

```python
all_movies
```

```python
movies_synopsys = pd.merge(all_movies, movie_data, on='imdb_title_id')
```

```python
movies_synopsys = movies_synopsys.drop(['imdb_title_id','len_description'], axis=1)
```

```python
movies_synopsys
```

```python
movies_synopsys.to_csv('synopsys.csv', header=None)
```

```python
with open('synopsys.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        print('movie pk: ', row[2])
        print('movie synopsys: ', row[1])
```
